```python
import pandas as pd
import numpy as np
import time
from selenium import webdriver
from collections import Counter
```

```python
browser = webdriver.Chrome()
url = "https://www.nadirkitap.com/kitap-ara.html"
browser.get(url)

x=0
linkler=[]
cevap = "Evet"
kitabevleri = []
kitaplistesi=[]
ücretsizlimit=[]
kargoücretleri=[]
fiyatlar=[]

dataframe=[]

while x<30 and cevap=="Evet":
    kitapadı=input("Kitap adı:")
    yazaradı=input("Yazar adı:")
    kitaplistesi.append(kitapadı)
    
    ja =browser.find_element_by_id("exampleInputFiyat")
    ja.send_keys("20")
    
    a=browser.find_element_by_id("exampleInputBookName")
    a.send_keys(kitapadı)
    
    b=browser.find_element_by_id("exampleInputYazar")
    b.send_keys(yazaradı)
    
    login = browser.find_element_by_css_selector('.btn.btn-default.btn-brd')
    login.click()
    
    #Kitabevi adlarını kaydediyoruz.
    c=browser.find_elements_by_xpath("//div/h4/a")
    links=[c1.get_attribute("href") for c1 in c]
    linkler.extend(links)
    
    for c1 in c:
        kitaplistesi.append(kitapadı)
    
    print("Başka kitap alacakmısınız?(Evet,Hayır)")
    cevap=input()
    x=x+1
    browser.get(url)

for i in linkler:
    browser.get(i)
    
    try:
        ö = browser.find_element_by_css_selector(".icon-tooltip-a")
        time.sleep(1)
        links=ö.get_attribute("data-original-title")
        kargoücretleri.append(links)
        
    except:
        print("1 yok")
        
        try:
            o = browser.find_element_by_css_selector(".icon-tooltip-y")
            time.sleep(1)
            links=o.get_attribute("data-original-title")
            kargoücretleri.append(links)
        except:
            kargoücretleri.append(" ")
    try:
        price = browser.find_element_by_css_selector(".product-price")
        links=price.text
        fiyatlar.append(links)
    except:
        fiyatlar.append(" ")

    try:
        t =browser.find_element_by_xpath("//div/p/span").text
        ücretsizlimit.append(t)
    except:
        ücretsizlimit.append(" ")
        
    ki = browser.find_element_by_xpath("//div/div/div/div/div/div/p/a")
    kii = ki.get_attribute("title")
    kitabevleri.append(kii)
```
